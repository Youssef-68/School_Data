```python
# import libraries
import piplite
await piplite.install('seaborn')

import pandas as pd
import numpy as np


import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import os
os.listdir()

# open the file
url = r"egypt_education_dataset.csv.zip"
df = pd.read_csv(url) #to loud file
df
```

```python
s_name = df['Student Name'].unique()
s_age = df['Student Age'].unique()
df.rename(columns= {'Student year':'Student Year'}, inplace= True)
s_year = df['Student Year'].unique()

f_degree = df['Father Degree'].unique()
m_degree = df['Mother Degree'].unique()
t_education = df['Education Type'].unique()

print(s_name, s_age, s_year, f_degree, m_degree, t_education)
```

```python
df['Total Grade'] = df['Subject_1'] + df['Subject_2'] + df['Subject_3'] + df['Subject_4'] + df['Subject_5'] + df['Subject_6'] + df['Subject_7'] + df['Subject_8'] + df['Subject_9'] + df['Subject_10']
df['Percentage'] = df['Total Grade'] * 0.1

df.fillna("Uncompleted")
```

```python
df['Student Age'] = pd.to_numeric(df['Student Age'], errors='coerce')

plt.figure(figsize=(6,6))
sns.histplot(df['Student Age'].dropna(), bins=12)
plt.ylim(9500, 10200)

plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.boxplot(x='Percentage', y='Education Type', data=df)

plt.title('Percentage Distribution by Education Type')
plt.xlabel('Percentage %')
plt.ylabel('Education Type')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Percentage', hue='Education Type', bins=45, kde=True, element='step') 

plt.title('Distribution of Percentage Scores by Education Type')
plt.xlabel('Percentage')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Father Degree', order=df['Father Degree'].value_counts().index)
plt.ylim(9000,10500)

plt.title('Count of Students by Father vs Education Level')
plt.xlabel('Father Degree')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Mother Degree', order=df['Mother Degree'].value_counts().index)
plt.ylim(9500,10200)

plt.title('Count of Students by Mother vs Education Level')
plt.xlabel('Mother Degree')
plt.ylabel('Count')
plt.show()
```
