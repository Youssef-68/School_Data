```python
# Importing Data Wrangling Libraries
import pandas as pd
import numpy as np

# Opening the dataset
url = r"/home/youssef/Desktop/Data_Projects/Egypt_Education/Data/egypt_education_dataset.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Making column names more consistent and easier to work with
new_columns = []

for col in df.columns:
    # Replace special characters with spaces
    clean_col = col.replace('-', ' ').replace('/', ' ').replace('.', ' ')
    # Split the cleaned column name into words
    words = clean_col.split()
    #  Capitalize each word and join them with underscores
    new_col = '_'.join(word.capitalize() for word in words)
    new_columns.append(new_col)

# Assign the new column names back to the DataFrame
df.columns = new_columns
```

```python
# Total number of rows
total_rows = len(df)
print("Total rows:", total_rows)

# Total number of columns
unique_names = df["Student_Name"].nunique()
print("Unique student names:", unique_names)
```

```python
# Count the occurrences of each name
name_counts = df["Student_Name"].value_counts()

# Names that appear more than once
duplicates = name_counts[name_counts > 1]
print(duplicates)
```

```python
# Check for fully duplicate rows
duplicate_rows = df[df.duplicated()]
print("Number of fully duplicate rows:", len(duplicate_rows))
```

```python
# Checking for missing values
df["Father_Degree"] = df["Father_Degree"].fillna("Unknown")
df["Mother_Degree"] = df["Mother_Degree"].fillna("Unknown")

# Extracting numeric part from "Student year" and converting it to integer
df["Student_Year"] = df["Student_Year"].astype(str)
df["Student_Year_num"] = df["Student_Year"].str.extract('(\d+)').astype(int)
```

```python
df.describe()
```

```python
# Creating new features for total and average scores, and categorizing performance
subjects = [f'Subject_{i}' for i in range(1, 11)]
df["Total_Score"] = df[subjects].sum(axis=1)
df["Average_Score"] = df[subjects].mean(axis=1)

# Function to categorize performance based on average score
def performance_category(avg):
    if avg >= 85: return "Excellent"
    elif avg >= 70: return "Good"
    elif avg >= 50: return "Average"
    else: return "Needs Improvement"

df["Performance"] = df["Average_Score"].apply(performance_category)
```

```python
# Saving the cleaned dataset for analysis
df_cleaned = df.copy()
df_cleaned.to_csv("/home/youssef/Desktop/Data_Projects/Egypt_Education/Data/education_cleaned.csv", index=False)
```
